```python
import gym
import numpy as np
import random
```

```python
env = gym.make("Taxi-v3").env
# Tabela Q 500 przykładów(przestrzeń obserwacji = 5*5*5*4 = 500) - 6 akcji (wschód, zachód, północ, południe, zabierz, wysadź)
q_table = np.zeros([env.observation_space.n,env.action_space.n])
```

```python
print(env.reset())
observation0 = env.reset()
print(observation0)
```

```python
alpha = 0.1 # współczynnik uczenia się
gamma = 0.9 # faktyczna zniżka
epsilon = 0.1 # współczynnik eksploracji
reward_list = [] # jest zdefiniowana, aby zachować nagrody
dropout_list = [] # jest zdefiniowana, aby zachować błędne odrzuty, aby zobaczyć jak działa algorytm
```

```python
episode_number = 5000

for i in range(1,episode_number):
    
    # init enviroment
    state = env.reset()
    
    reward_count = 0
    dropouts = 0
    
    while True:
        
        # exploit vs explore to find action epsilon 0.1 => %10 explore %90 explotit
        if random.uniform(0,1) < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(q_table[state])
            
        # action process and take reward / take observation
        next_state, reward, done, _ = env.step(action)
        
        # q learning funct
        
        old_value = q_table[state, action]  #old value
        next_max = np.max(q_table[next_state]) #next max
        
        next_value = (1-alpha)*old_value + alpha*(reward + gamma*next_max)
        # q table update
        q_table[state,action] = next_value
        
        # update state
        state = next_state
        
        # find wrong dropout 
        if reward == -10:
            dropouts += 1
            
        if done:
            break
        
        reward_count  += reward
    if i%10 == 0:
        
        dropout_list.append(dropouts)
        reward_list.append(reward_count)
        print("Episode: {}, reward {}, wrong dropout {}".format(i, reward_count,dropouts))
```

```python
env.s = env.encode(0,0,3,4) # env.encode(taxi row, taxi col, passenger index, destination)
env.render()
```
